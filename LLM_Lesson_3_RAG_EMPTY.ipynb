# Lexical Search vs Embedding Search

## Lexical Search

```python
!pip install rank_bm25 -qU
!pip show rank_bm25  #0.2.2
```

```python
from rank_bm25 import BM25Okapi
import os
import numpy as np
```

```python
def create_bm25_index(chunks):
    """Create BM25 index from contextualized chunks."""
    tokenized_chunks = [chunk.split() for chunk in chunks]
    return BM25Okapi(tokenized_chunks)
```

```python
def chunk_document(document, chunk_size=200, overlap=50):
    """Split document into chunks of specified size with overlap."""
    words = document.split()
    chunks = []
    start = 0
    while start < len(words):
        end = min(start + chunk_size, len(words))
        chunk = " ".join(words[start:end])
        chunks.append(chunk)
        start += chunk_size - overlap
    return chunks
```

```python
# Function to read text from a PDF
def read_pdf(file_path):
    pages_txt = []
    with open(file_path, 'rb') as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            pages_txt.append(page.extract_text() or '')
    return pages_txt
```

```python
import PyPDF2
pdf_dir = "/content/acct725_teachingmaterials/pdfs/"
documents = []
for file in os.listdir(pdf_dir):
    if file.endswith(".pdf"):
        documents.extend(read_pdf(os.path.join(pdf_dir, file)))
```

```python
len(documents)
```

```python
all_chunks = []
chunk_to_doc_mapping = []  # Track which document each chunk belongs to

print("Processing documents...")
for i, doc in enumerate(documents):
    # print(f"Processing document {i+1}/{len(documents)}")
    chunks = chunk_document(doc)
    all_chunks.extend(chunks)
    chunk_to_doc_mapping.extend([i] * len(chunks))
```

```python

bm25_index = create_bm25_index(all_chunks)
```

```python
def retrieve_chunks(query, scores, chunks, top_k=20):
    """Retrieve top-k relevant chunks using scores."""

    top_indices = np.argsort(scores)[::-1][:top_k]

    # Return top chunks
    return [chunks[i] for i in top_indices]
```

```python
query = "Explain why Netflix is interested to acquire Next Games?"
bm25_scores = bm25_index.get_scores(query.split())
relevant_chunks_lex = retrieve_chunks(query,
                bm25_scores,
                chunks = all_chunks,
                top_k = 10)
```

## Step 2: Load PDFs & Create Vector Database

```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
import os
```

```python
# Load all PDFs
pdf_dir = "/content/acct725_teachingmaterials/pdfs/"
documents = []
for file in os.listdir(pdf_dir):
    if file.endswith(".pdf"):
        loader = PyPDFLoader(os.path.join(pdf_dir, file))
        documents.extend(loader.load())

# Split text into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
texts = text_splitter.split_documents(documents)

# Create embeddings and FAISS index
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = FAISS.from_documents(texts, embeddings)

# Save to disk (optional)
vectorstore.save_local("faiss_index")
```

## Step 3: Initialize LLM & QA Chain

```python
from langchain.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA

from langchain.chat_models import ChatOpenAI

llm = ChatOpenAI(model_name="llama-3.1-8b-instant", temperature=0,
                 base_url="https://api.groq.com/openai/v1",
                 api_key=os.environ.get("GROQ_API_KEY")
                 )
```

```python
# Create retriever
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 20})

# Build QA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
```

## Step 4: Ask a Question

```python
def ask_question(question):
    result = qa_chain({"query": question})
    print("Answer:", result["result"])
    print("\nSource Documents:")
    for doc in result["source_documents"]:
        print(f"- {doc.metadata['source']}: {doc.page_content[:200]}...")

    return result["result"]

# Test it!
ans = ask_question("Which region is netflix expecting the most growth?")
```

```python
# 1. Load PDF documents
from langchain_community.document_loaders import DirectoryLoader
from ragas.testset import TestsetGenerator
from ragas.llms import LangchainLLMWrapper
from ragas.embeddings import LangchainEmbeddingsWrapper
from langchain_groq import ChatGroq
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.retrievers import BM25Retriever
from langchain.chains import RetrievalQA
from ragas.metrics import faithfulness, answer_relevancy, context_precision, context_recall
from ragas import evaluate
from datasets import Dataset
import os

loader = DirectoryLoader("/content/acct725_teachingmaterials/pdfs/", glob="**/*Q4 2023*.pdf")
docs = loader.load_and_split()
```

```python
# 2. Set up LLM and embeddings for test generation
groq_api_key = os.getenv("GROQ_API_KEY")
llm = ChatGroq(model="llama3-8b-8192", api_key=groq_api_key)
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Wrap your models
wrapped_llm = LangchainLLMWrapper(llm)
wrapped_embeddings = LangchainEmbeddingsWrapper(embedding_model)

generator = TestsetGenerator(
    llm=wrapped_llm,
    embedding_model=wrapped_embeddings
)

testset = generator.generate_with_langchain_docs(
    documents=docs,
    testset_size=5,
)

# Convert to evaluation dataset (will be updated later with RAG responses)
testset_df = testset.to_pandas()
testset_df.columns
```

```python
from ragas import EvaluationDataset

bm25_retriever = BM25Retriever.from_documents(docs)
bm25_retriever.k = 3

lexical_rag = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=bm25_retriever
)
vectorstore = FAISS.from_documents(docs, embedding_model)
embedding_retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

embedding_rag = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=embedding_retriever
)

def format_docs(relevant_docs): return "\n".join(doc.page_content for doc in relevant_docs)
```

```python
from ragas import evaluate
from ragas.llms import LangchainLLMWrapper
from ragas.metrics import LLMContextRecall, Faithfulness, FactualCorrectness
from langchain_openai import ChatOpenAI

def evaluate_rag(retriever, rag):
  dataset = []
  for _, row in testset_df.iterrows():
      query = row['user_input']

      # Run your RAG system
      relevant_docs = retriever.invoke(query)
      rag_result = rag.invoke({"query": query})  # Remove context parameter

      # Extract the string response from the result
      response = rag_result["result"] if isinstance(rag_result, dict) else rag_result

      dataset.append({
          "user_input": query,
          "retrieved_contexts": [doc.page_content for doc in relevant_docs],
          "response": response,  # Now this is a string
          "reference": row.get('reference', '')
      })

  evaluation_dataset = EvaluationDataset.from_list(dataset)
  evaluator_llm = LangchainLLMWrapper(llm)

  # Run evaluation
  result = evaluate(
      dataset=evaluation_dataset,
      metrics=[LLMContextRecall(), Faithfulness(), FactualCorrectness()],
      llm=evaluator_llm,
  )

  print(result)
  # Convert to pandas for analysis
  result_df = result.to_pandas()

  return result, result_df
```

```python
import os
import numpy as np
from rank_bm25 import BM25Okapi
from sentence_transformers import SentenceTransformer
# from anthropic import Anthropic

# Initialize models and clients
# anthropic_client = Anthropic(api_key=os.getenv("ANTHROPIC_API_KEY"))

embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
from groq import Groq
client = Groq()

def chunk_document(document, chunk_size=200, overlap=50):
    """Split document into chunks of specified size with overlap."""
    words = document.split()
    chunks = []
    start = 0
    while start < len(words):
        end = min(start + chunk_size, len(words))
        chunk = " ".join(words[start:end])
        chunks.append(chunk)
        start += chunk_size - overlap
    return chunks

def generate_context(document, chunk):
    """Generate concise context for a chunk using groq."""
    prompt = f"<document>\n{document}\n</document>\nHere is the chunk we want to situate within the whole document\n<chunk>\n{chunk}\n</chunk>\nPlease give a short succinct context to situate this chunk within the overall document for the purposes of improving search retrieval of the chunk. Answer only with the succinct context and nothing else."
    response = client.chat.completions.create(
        # model="claude-3-haiku",
        model="llama3-8b-8192",
        messages=[
    {
        "role": "user",
        "content": (
            f"{prompt}"
        ),
    },
],
        max_tokens=100
    )
    return response.choices[0].message.content

def create_contextualized_chunks(document):
    """Create chunks and prepend generated context to each."""
    chunks = chunk_document(document)
    contextualized_chunks = []
    for chunk in chunks:
        context = generate_context(document, chunk)
        contextualized_chunk = f"{context} {chunk}"
        contextualized_chunks.append(contextualized_chunk)
    return contextualized_chunks

def create_embeddings(chunks):
    """Generate vector embeddings for contextualized chunks."""
    return embedding_model.encode(chunks)

def create_bm25_index(chunks):
    """Create BM25 index from contextualized chunks."""
    tokenized_chunks = [chunk.split() for chunk in chunks]
    return BM25Okapi(tokenized_chunks)

def retrieve_chunks(query, embeddings, bm25, contextualized_chunks, top_k=20):
    """Retrieve top-k relevant chunks using embeddings and BM25."""
    # Convert query to embedding
    query_embedding = embedding_model.encode([query])[0]

    # Calculate embedding similarity scores
    embedding_scores = np.dot(embeddings, query_embedding)

    # Calculate BM25 scores
    bm25_scores = bm25.get_scores(query.split())

    # Combine scores (simple addition as a basic rank fusion)
    combined_scores = embedding_scores + bm25_scores

    # Get top-k indices
    top_indices = np.argsort(combined_scores)[::-1][:top_k]

    # Return top chunks
    return [contextualized_chunks[i] for i in top_indices]
```

```python
def process_documents(documents):
    """Process multiple documents and create retrieval indices."""
    all_contextualized_chunks = []
    chunk_to_doc_mapping = []  # Track which document each chunk belongs to

    print("Processing documents...")
    for i, doc in enumerate(documents):
        print(f"Processing document {i+1}/{len(documents)}")
        contextualized_chunks = create_contextualized_chunks(doc)
        all_contextualized_chunks.extend(contextualized_chunks)
        chunk_to_doc_mapping.extend([i] * len(contextualized_chunks))

    print("Creating embeddings...")
    embeddings = create_embeddings(all_contextualized_chunks)

    print("Creating BM25 index...")
    bm25_index = create_bm25_index(all_contextualized_chunks)

    return all_contextualized_chunks, embeddings, bm25_index, chunk_to_doc_mapping
```

```python
pdf_dir = "/content/acct725_teachingmaterials/pdfs/"
documents = []

for file in os.listdir(pdf_dir):
    if file.endswith(".pdf"):
        loader = PyPDFLoader(os.path.join(pdf_dir, file))
        pages = loader.load()
        full_text = "\n".join([page.page_content for page in pages])[:10_000] #truncate the pages to save time/token spent
        documents.append(full_text)
```

```python
def process_documents(documents):
    """
    Process a list of documents into contextualized chunks.

    Args:
        documents (list of str): List of documents to process.

    Returns:
        list of str: List of contextualized chunks from all documents.
    """
    all_contextualized_chunks = []
    for i, document in enumerate(documents):
        print(f'Chunking {i+1}/{len(documents)}')
        chunks = create_contextualized_chunks(document)
        all_contextualized_chunks.extend(chunks)
    return all_contextualized_chunks

def generate_answer(query, top_chunks):
    """
    Generate an answer to the query based on the top retrieved chunks using Groq API.

    Args:
        query (str): The question to answer.
        top_chunks (list of str): List of top retrieved contextualized chunks.

    Returns:
        str: The generated answer.
    """
    prompt = f"Based on the following information, answer the question: {query}\n\n"
    for i, chunk in enumerate(top_chunks, 1):
        prompt += f"Information {i}:\n{chunk}\n\n"
    prompt += "Answer:"

    response = client.chat.completions.create(
        model="llama3-8b-8192",
        messages=[
            {
                "role": "user",
                "content": prompt,
            },
        ],
        max_tokens=500
    )
    return response.choices[0].message.content

# Example usage:
contextualized_chunks = process_documents(documents)
embeddings = create_embeddings(contextualized_chunks)
bm25 = create_bm25_index(contextualized_chunks)
```

```python
query = "Which region is netflix expecting the most growth?"
top_chunks = retrieve_chunks(query, embeddings, bm25, contextualized_chunks, top_k=20)
answer = generate_answer(query, top_chunks)
display(Markdown(answer))
```
